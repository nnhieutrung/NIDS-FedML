```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("IoTiD20.zip")
```

```python
df
```

```python
output = 'Cat'
```

```python
df[output].unique()
```

```python
df[output].value_counts()
```

```python
col_1 = ['Scan','Mirai']
col_2 = ['DoS', 'MITM ARP Spoofing']
```

```python
df_1 = df[df[output].isin(col_1)] 
df_2 = df[df[output].isin(col_2)]
df_3 = df[df[output] == 'Normal']
```

```python
midpoint = len(df_3) // 2

df_1 = pd.concat([df_1, df_3.iloc[:midpoint]], axis=False)
df_2 = pd.concat([df_2, df_3.iloc[:midpoint]], axis=False)
```

```python
df_1[output].count(), df_2[output].count()
```

```python
df_1[output].value_counts()
```

```python
df_2[output].value_counts()
```

```python
def csv_save_zip(df, filename):
    compression_options = dict(method='zip', archive_name=f'{filename}.csv')
    df.to_csv(f'{filename}.zip', compression=compression_options, index=False)
```

```python
csv_save_zip(df_1, 'training-set_0')
```

```python
csv_save_zip(df_2, 'training-set_1')
```

```python
x = df
y = df[output]
x, _x, y, _y = train_test_split(x, y, random_state=189, test_size=0.05, stratify=y)
```

```python
csv_save_zip(x, 'training-set_0')
```
